```python
data = df[['Lable',
       'Q.1 The content was as described in publicity materials',    
       'Q.2 The workshop was applicable to my job',
       "Q.3 I'll recommend this workshop to other conservators.",
       'Q.4 The program was well paced within the allotted time',
       'Q.5 The instructor was a good communicator ',
       'Q.6 The material was presented in an organized manner',
       'Q.7 The instructor was knowledgeable on topics ',
       "Q.8 I'd be interested in tending a followup, more advanced workshop on the same subject",
       'Q.9 Given the topic, was this workshop',
       'Q.10 In your opinion, was this workshop',
       'Q.11 Please Rate the following : Visual',
       'Q.12 Please Rate the following : Acoustics',
       'Q.13 Please Rate the following : Meeting Space',
       'Q.14 Please Rate the following : Handouts',
       'Q.15 Please Rate the following : The program overall','score']]
```

```python
data.head()
```

```python
from skleargn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('Lable',axis=1), data['Lable'], test_size=0.33, random_state =101)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train , y_train)
```

```python
predicted= model.predict(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
print(confusion_matrix(y_test , predicted))
```

```python
print(classification_report(y_test , predicted))
```
